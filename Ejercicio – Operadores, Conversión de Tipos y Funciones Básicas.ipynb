```python
# Solicitar al usuario la operación
operacion = input("Seleccione una operación (suma, resta, multiplicacion, division o potencia): ")

# Solicitar al usuario ingresar una lista de 5 números
entrada_lista = input("Ingrese una lista de 5 números separados por comas (por ejemplo, 1,2,3,4,5): ")
entrada_lista = entrada_lista.split(',')
try:
    entrada_lista = [float(num) for num in entrada_lista]
except ValueError:
    entrada_lista = []

# Solicitar al usuario ingresar un conjunto (set) de 5 números
entrada_set = input("Ingrese un conjunto (set) de 5 números separados por comas (por ejemplo, 1,2,3,4,5): ")
entrada_set = entrada_set.split(',')
try:
    entrada_set = set(float(num) for num in entrada_set)
except ValueError:
    entrada_set = set()

# Solicitar al usuario ingresar un diccionario de 5 pares clave-valor
entrada_diccionario = {}
for i in range(5):
    clave = input(f"Ingrese la clave {i + 1} para el diccionario: ")
    valor = input(f"Ingrese el valor {i + 1} para el diccionario: ")
    try:
        valor = float(valor)
    except ValueError:
        valor = 0
    entrada_diccionario[clave] = valor

# Realizar operaciones y mostrar resultados
if len(entrada_lista) == 5:
    print("\nEntrada de tipo lista válida:", entrada_lista)
    if operacion == 'suma':
        resultado = sum(entrada_lista)
    elif operacion == 'resta':
        resultado = entrada_lista[0] - sum(entrada_lista[1:])
    elif operacion == 'multiplicacion':
        resultado = 1
        for num in entrada_lista:
            resultado *= num
    elif operacion == 'division':
        resultado = entrada_lista[0]
        for num in entrada_lista[1:]:
            resultado /= num
    elif operacion == 'potencia':
        resultado = entrada_lista[0]
        for num in entrada_lista[1:]:
            resultado **= num
    else:
        resultado = None

    if resultado is not None:
        print(f"Resultado de la operación '{operacion}': {resultado}")
    else:
        print("Operación no válida.")
else:
    print("\nEntrada de tipo lista inválida.")

if len(entrada_set) == 5:
    print("\nEntrada de tipo set válida:", entrada_set)
    lista_set = list(entrada_set)
    if operacion == 'suma':
        resultado = sum(lista_set)
    elif operacion == 'resta':
        resultado = lista_set[0] - sum(lista_set[1:])
    elif operacion == 'multiplicacion':
        resultado = 1
        for num in lista_set:
            resultado *= num
    elif operacion == 'division':
        resultado = lista_set[0]
        for num in lista_set[1:]:
            resultado /= num
    elif operacion == 'potencia':
        resultado = lista_set[0]
        for num in lista_set[1:]:
            resultado **= num
    else:
        resultado = None

    if resultado is not None:
        print(f"Resultado de la operación '{operacion}': {resultado}")
    else:
        print("Operación no válida.")
else:
    print("\nEntrada de tipo set inválida.")

if len(entrada_diccionario) == 5:
    print("\nEntrada de tipo diccionario válida:", entrada_diccionario)
    valores_diccionario = list(entrada_diccionario.values())
    if operacion == 'suma':
        resultado = sum(valores_diccionario)
    elif operacion == 'resta':
        resultado = valores_diccionario[0] - sum(valores_diccionario[1:])
    elif operacion == 'multiplicacion':
        resultado = 1
        for num in valores_diccionario:
            resultado *= num
    elif operacion == 'division':
        resultado = valores_diccionario[0]
        for num in valores_diccionario[1:]:
            resultado /= num
    elif operacion == 'potencia':
        resultado = valores_diccionario[0]
        for num in valores_diccionario[1:]:
            resultado **= num
    else:
        resultado = None

    if resultado is not None:
        print(f"Resultado de la operación '{operacion}': {resultado}")
    else:
        print("Operación no válida.")
else:
    print("\nEntrada de tipo diccionario inválida.")
```
